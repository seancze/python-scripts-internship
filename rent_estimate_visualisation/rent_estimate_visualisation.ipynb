```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

```python
def plot_bell_curve(filepath):
    with open(filepath) as f:
        output = json.load(f)

        out_x = [el[0] for el in output['bell_curve']]
        out_y = [el[1] for el in output['bell_curve']]

        # Plotting the bell-shaped curve
        plt.style.use('seaborn')
        plt.figure(figsize = (6, 6))
        plt.plot(out_x, out_y, color = 'black',
                 linestyle = 'dashed')

        plt.scatter( out_x, out_y, marker = 'o', s = 25, color = 'red')
        plt.show()
```

```python
def plot_histogram(filepath):
    with open(filepath) as f:
        output = json.load(f)
        
        plt.hist(output['histogram'], density=True, bins=30)  # density=False would make counts
        plt.ylabel('Probability')
        plt.xlabel('Data');
```

## Multiple neighbourhoods

```python
multiple_neighbourhoods_filepath = 'data/out_multiple.json'
single_neighbourhood_filepath = 'data/out_single.json'

plot_bell_curve(multiple_neighbourhoods_filepath)
plot_histogram(multiple_neighbourhoods_filepath)
```

## Single neighbourhood

```python
plot_bell_curve(single_neighbourhood_filepath)
plot_histogram(single_neighbourhood_filepath)
```
